This notebook demonstrates a computational approach to the Collatz conjecture,
searching for potential counterexamples and analyzing the trajectory statistics.

```python
def collatz_steps(n: int) -> int:
    """Count steps to reach 1 in the Collatz sequence."""
    steps = 0
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        steps += 1
    return steps

def collatz_max(n: int) -> int:
    """Find the maximum value in the Collatz sequence starting from n."""
    max_val = n
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        max_val = max(max_val, n)
    return max_val
```

```python
# Verify Collatz for a range of integers
N = 100_000
print(f'Verifying Collatz conjecture for n = 1 to {N:,}...')

max_steps = 0
max_steps_n = 1
max_height = 0
max_height_n = 1

for n in range(1, N + 1):
    steps = collatz_steps(n)
    height = collatz_max(n)
    
    if steps > max_steps:
        max_steps = steps
        max_steps_n = n
    if height > max_height:
        max_height = height
        max_height_n = n

print(f'All integers 1..{N:,} converge to 1.')
print(f'Max steps: {max_steps} (at n={max_steps_n:,})')
print(f'Max height: {max_height:,} (at n={max_height_n:,})')
```

```python
# Analyze step count distribution
step_counts = [collatz_steps(n) for n in range(1, N + 1)]

import statistics
print(f'Step count statistics for n = 1..{N:,}:')
print(f'  Mean:   {statistics.mean(step_counts):.1f}')
print(f'  Median: {statistics.median(step_counts):.1f}')
print(f'  Stdev:  {statistics.stdev(step_counts):.1f}')
print(f'  Max:    {max(step_counts)}')
```

```python
# Look for "record-breaking" numbers — integers with unusually many steps
records = []
best = 0
for n in range(1, N + 1):
    s = collatz_steps(n)
    if s > best:
        best = s
        records.append((n, s))

print(f'Record-breaking step counts (n, steps):')
for n, s in records[-10:]:
    print(f'  n={n:>8,}  steps={s}')
```

## Limitations

- This is **computational evidence only**, not a proof.
- The conjecture has been verified for all integers up to ~2^68 by others.
- Our search covers only up to 100,000 (or whatever N is set to).
- A counterexample, if one exists, could be astronomically large.
