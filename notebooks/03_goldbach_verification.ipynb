This notebook verifies Goldbach's conjecture for even numbers up to a given bound
and analyzes the number of representations as a sum of two primes.

```python
def sieve_of_eratosthenes(limit: int) -> list[bool]:
    """Return a boolean sieve where is_prime[i] is True if i is prime."""
    is_prime = [True] * (limit + 1)
    is_prime[0] = is_prime[1] = False
    for i in range(2, int(limit**0.5) + 1):
        if is_prime[i]:
            for j in range(i*i, limit + 1, i):
                is_prime[j] = False
    return is_prime
```

```python
LIMIT = 100_000
print(f'Generating prime sieve up to {LIMIT:,}...')
is_prime = sieve_of_eratosthenes(LIMIT)
primes = [i for i in range(2, LIMIT + 1) if is_prime[i]]
print(f'Found {len(primes):,} primes up to {LIMIT:,}')
```

```python
def goldbach_representations(n: int, is_prime: list[bool]) -> int:
    """Count the number of ways n can be written as p + q with p <= q both prime."""
    count = 0
    for p in range(2, n // 2 + 1):
        if is_prime[p] and is_prime[n - p]:
            count += 1
    return count
```

```python
# Verify Goldbach's conjecture for all even numbers from 4 to LIMIT
print(f'Verifying Goldbach\'s conjecture for even numbers 4 to {LIMIT:,}...')

min_reps = float('inf')
min_reps_n = 4
max_reps = 0
max_reps_n = 4
failures = []

for n in range(4, LIMIT + 1, 2):
    reps = goldbach_representations(n, is_prime)
    if reps == 0:
        failures.append(n)
    if reps < min_reps:
        min_reps = reps
        min_reps_n = n
    if reps > max_reps:
        max_reps = reps
        max_reps_n = n

if failures:
    print(f'COUNTEREXAMPLES FOUND: {failures[:10]}')
else:
    print(f'Conjecture verified for all {(LIMIT - 2) // 2:,} even numbers from 4 to {LIMIT:,}.')
    print(f'Min representations: {min_reps} (at n={min_reps_n})')
    print(f'Max representations: {max_reps} (at n={max_reps_n:,})')
```

```python
# Analyze the "Goldbach comet" — number of representations vs n
sample_points = list(range(4, min(LIMIT + 1, 10_001), 2))
reps_data = [(n, goldbach_representations(n, is_prime)) for n in sample_points]

# Find numbers with fewest representations (hardest to decompose)
reps_data_sorted = sorted(reps_data, key=lambda x: x[1])
print('Even numbers with fewest Goldbach representations:')
for n, r in reps_data_sorted[:10]:
    print(f'  {n}: {r} representation(s)')
```

## Limitations
